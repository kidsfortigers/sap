```python
import pandas as pd
import numpy as np
```

```python
current_price = np.array([np.random.randint(0,101) for i in np.arange(0,100)])
current_price
```

```python
one_week_movement = np.array([np.random.randint(-5,6) for i in np.arange(0,100)])
one_week_movement
```

```python
in_the_news = np.array([np.random.randint(0,2) for i in np.arange(0,100)])
in_the_news
```

```python
sentiment_twitter = np.array([np.random.randint(0,3) for i in np.arange(0,100)])
sentiment_twitter
```

```python
table_of_strategy = pd.DataFrame({ 'current_price' : current_price, 'one_week_movement' : one_week_movement,
                                  'news' : in_the_news, 'twitter': sentiment_twitter })
table_of_strategy
```

```python
# Keys assumed for the next bit
# Movement is in percentages 
# in_the_news: 0 is no, 1 is yes
# twitter: 0 is neg, 1 is pos, 2 is neutral/not significant
# 1 is move, 0 is stay
```

```python
output_class = []
```

```python
for index in np.arange(0, 100):
    if one_week_movement[index] >= 0:
        if sentiment_twitter[index] == 0:
            if in_the_news[index] == 0:
                output_class = output_class + [0]
#                 Gained, twitter hate, not in news - Uncertainity
            else:
                output_class = output_class + [0]
#                 Gained, twitter hate, in news - Mass Movement - Soft 1
        else:
            if in_the_news[index] == 0:
                output_class = output_class + [1]
#                Gained, twitter love, not in new - Move
            else:
                output_class = output_class + [0]
#             Gained, twitter love, in news - Mass Movement
    else:
        if sentiment_twitter[index] == 0:
            if in_the_news[index] == 0:
                output_class = output_class + [1]
#                 Lost, twitter hate, not in news - Move
            else:
                output_class = output_class + [0]
#             Lost, twitter hate, in news - Mass Movement
        else:
            if in_the_news[index] == 0:
                output_class = output_class + [0]
#                 Lost, twitter love, not in news - Uncertainity
            else:
                output_class = output_class + [1]
#             Lost, twitter love, in news - Mass Movement
```

```python
out_class = np.array(output_class)
out_class
```

```python
table_of_strategy = pd.DataFrame({ 'current_price' : current_price, 'one_week_movement' : one_week_movement,
                                  'news' : in_the_news, 'twitter': sentiment_twitter, 'class': out_class })
table_of_strategy
```

```python
table_of_strategy.to_csv('train_data_attempt2')
```
